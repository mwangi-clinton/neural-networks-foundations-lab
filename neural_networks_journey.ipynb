## 1. The Foundation: The Perceptron
The perceptron is the simplest form of a neural network, consisting of a single node. It takes inputs, multiplies them by weights, adds a bias, and passes the result through an activation function.

### Mathematical Formulation
Given an input vector $\mathbf{x} = [x_1, x_2, \dots, x_n]$, a weight vector $\mathbf{w} = [w_1, w_2, \dots, w_n]$, and a bias $b$, the pre-activation output $z$ is:

$$ z = \sum_{i=1}^{n} w_i x_i + b = \mathbf{w}^T \mathbf{x} + b $$

The final output $\hat{y}$ is obtained by applying an activation function $f(z)$:
$$ \hat{y} = f(z) $$

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F

class Perceptron:
    def __init__(self, input_size):
        # Initialize weights and bias randomly
        self.weights = np.random.randn(input_size)
        self.bias = np.random.randn()
        
    def step_function(self, z):
        # Simple threshold activation function
        return 1 if z >= 0 else 0
        
    def forward(self, x):
        # Linear combination: z = w^T * x + b
        z = np.dot(self.weights, x) + self.bias
        return self.step_function(z)

# Example usage
x = np.array([0.5, -0.2])
model = Perceptron(input_size=2)
print(f"Inputs: {x}")
print(f"Weights: {model.weights}")
print(f"Bias: {model.bias}")
print(f"Output: {model.forward(x)}")
```

## 2. Activation Functions
Activation functions introduce non-linearities into the network, allowing it to learn complex, non-linear patterns. Without them, a neural network, no matter how deep, would just be a linear transformation.

### Common Functions:
1. **Sigmoid**: S-shaped curve, outputs between 0 and 1. Prone to vanishing gradients.
   $$ \sigma(x) = \frac{1}{1 + e^{-x}} $$
2. **Hyperbolic Tangent (Tanh)**: Outputs between -1 and 1. Zero-centered.
   $$ \text{tanh}(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} $$
3. **Rectified Linear Unit (ReLU)**: Most popular for hidden layers. Avoids vanishing gradient for positive values.
   $$ \text{ReLU}(x) = \max(0, x) $$
4. **Softmax**: Converts a vector of numbers into a probability distribution (used in multi-class classification).
   $$ \text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_j e^{x_j}} $$

```python
x_vals = np.linspace(-5, 5, 100)

def sigmoid(x): return 1 / (1 + np.exp(-x))
def tanh(x): return np.tanh(x)
def relu(x): return np.maximum(0, x)

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.plot(x_vals, sigmoid(x_vals), label="Sigmoid")
plt.title("Sigmoid"); plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(x_vals, tanh(x_vals), label="Tanh", color="orange")
plt.title("Tanh"); plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(x_vals, relu(x_vals), label="ReLU", color="green")
plt.title("ReLU"); plt.grid(True)

plt.tight_layout()
plt.show()
```

## 3. Backpropagation and Automatic Differentiation
Training a neural network requires adjusting the weights to minimize a loss function $L$. This is done using **Gradient Descent**:

$$ w_{new} = w_{old} - \eta \frac{\partial L}{\partial w} $$
where $\eta$ is the learning rate.

**Backpropagation** uses the Chain Rule of Calculus to efficiently compute these gradients across all layers. 
For a composition of functions $y = f(g(x))$, the chain rule is:
$$ \frac{dy}{dx} = \frac{dy}{dg} \cdot \frac{dg}{dx} $$

Modern deep learning frameworks use **Automatic Differentiation (Autograd)** to track operations dynamically and compute gradients automatically without manual calculus.

```python
# Automatic Differentiation with PyTorch
# Let's define a simple computation graph: y = (x * w + b)^2

x = torch.tensor([2.0], requires_grad=False)
w = torch.tensor([3.0], requires_grad=True)  # We want to optimize w, so requires_grad=True
b = torch.tensor([1.0], requires_grad=True)  # We want to optimize b, so requires_grad=True

# Forward pass
z = x * w + b
y = z ** 2

print(f"Forward pass output (y): {y.item()}")

# Backward pass (Automatic Differentiation)
y.backward()

# Analytical Gradients for verification:
# dy/dw = 2 * (x * w + b) * x = 2 * (2*3 + 1) * 2 = 28
# dy/db = 2 * (x * w + b) * 1 = 2 * (2*3 + 1) * 1 = 14

print(f"Gradient of w (dy/dw): {w.grad.item()}")
print(f"Gradient of b (dy/db): {b.grad.item()}")
```

## 4. Advanced Architectures: Convolutional Neural Networks (CNNs)
CNNs are specialized for processing grid-like topology data, such as images. They use convolution operations instead of standard matrix multiplications in at least one layer.

### The Convolution Operation
A convolution slides a small matrix (the **kernel** or **filter**) over the input to produce a **feature map**. For a 2D input $I$ and kernel $K$:

$$ S(i, j) = (I * K)(i, j) = \sum_m \sum_n I(i+m, j+n) K(m, n) $$

This allows the network to learn translation-invariant spatial features (edges, textures, shapes).

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        # 1 input channel (e.g., grayscale image), 16 output channels, 3x3 kernel
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        # Assuming 28x28 input (e.g. MNIST), after one 2x2 pooling, size becomes 14x14
        self.fc1 = nn.Linear(16 * 14 * 14, 10) 

    def forward(self, x):
        # Input -> Conv -> ReLU -> Pool
        x = self.pool(F.relu(self.conv1(x)))
        x = x.view(-1, 16 * 14 * 14) # Flatten the tensor for the fully connected layer
        x = self.fc1(x)
        return x

cnn_model = SimpleCNN()
dummy_image = torch.randn(1, 1, 28, 28) # Batch size 1, 1 channel, 28x28 pixels
output = cnn_model(dummy_image)
print(f"CNN Output shape: {output.shape}") # Should be [1, 10] (e.g., 10 classes)
```

## 5. Advanced Architectures: Transformers
Transformers revolutionized sequence modeling (like Natural Language Processing) by dropping recurrence (RNNs) entirely and relying purely on the **Attention Mechanism**.

### Scaled Dot-Product Attention
The core of the Transformer is Self-Attention. It computes an output based on three matrices derived from the input sequence: **Queries ($Q$)**, **Keys ($K$)**, and **Values ($V$)**.

$$ \text{Attention}(Q, K, V) = \text{softmax}\left( \frac{QK^T}{\sqrt{d_k}} \right) V $$

Where $d_k$ is the dimension of the key vectors. The dot product $QK^T$ computes similarity scores between elements in the sequence, which are normalized by the softmax function to become attention weights. These weights decide how much "attention" each element in the sequence should pay to other elements.

```python
class SelfAttention(nn.Module):
    def __init__(self, embed_size):
        super(SelfAttention, self).__init__()
        self.embed_size = embed_size
        
        # Linear layers to project input into Q, K, V
        self.queries = nn.Linear(embed_size, embed_size, bias=False)
        self.keys = nn.Linear(embed_size, embed_size, bias=False)
        self.values = nn.Linear(embed_size, embed_size, bias=False)
        
    def forward(self, x):
        # x shape: (batch_size, sequence_length, embed_size)
        Q = self.queries(x)
        K = self.keys(x)
        V = self.values(x)
        
        # Compute Dot Product: QK^T
        # Q: (B, SeqLen, E) | K.transpose: (B, E, SeqLen) -> Score: (B, SeqLen, SeqLen)
        attention_scores = torch.bmm(Q, K.transpose(1, 2)) / (self.embed_size ** 0.5)
        
        # Softmax over the last dimension to get weights
        attention_weights = F.softmax(attention_scores, dim=-1)
        
        # Multiply by V
        # weights: (B, SeqLen, SeqLen) | V: (B, SeqLen, E) -> Out: (B, SeqLen, E)
        out = torch.bmm(attention_weights, V)
        return out, attention_weights

# Testing the Self-Attention mechanism
seq_len = 5
embed_size = 64
attention_layer = SelfAttention(embed_size)

dummy_sequence = torch.randn(1, seq_len, embed_size) # Batch size 1, sequence of 5 tokens

out, weights = attention_layer(dummy_sequence)
print(f"Input sequence shape: {dummy_sequence.shape}")
print(f"Attention weights shape: {weights.shape} (5 tokens attending to 5 tokens)")
print(f"Output sequence shape: {out.shape}")
```
